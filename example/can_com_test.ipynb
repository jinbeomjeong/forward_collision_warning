```python
import time, can, sys, os
import numpy as np
import pandas as pd

sys.path.append(os.getcwd()+'\\utils')
```

```python
from utils.system_communication import AdasCANCommunication
```

```python
can_channel = can.interface.Bus(bustype='pcan', channel='PCAN_USBBUS1', bitrate=500000)
```

```python
data = pd.read_csv('../data/timing_chart.csv')
data
```

```python
adas_msg_parser = AdasCANCommunication()
```

```python
time_val = time.time()
state_of_dsm_val: int = 0
state_of_fcw_val: int = 0
object_type_val: int = 0
ttc_val: float = 0.0
state_of_ldws_val: int = 0
```

```python
t0 = time.time()
adas_msg = []

while True:
    tc = time.time() - t0
    
    sim_index = np.argmin(np.abs(data['time(sec)'].to_numpy()- tc))
    sim_value = data.iloc[sim_index]
    
    time_val = float(sim_value['time(sec)'].item())
    state_of_dsm_val = int(sim_value['state_of_DSM'].item())
    state_of_fcw_val = int(sim_value['state_of_FCW'].item())
    object_type_val = int(sim_value['object_type'].item())
    ttc_val = float(sim_value['time_to_collision(sec)'].item())
    state_of_ldws_val = int(sim_value['state_of_LDWS'].item())
    
    adas_msg.append(adas_msg_parser.create_dsm_can_msg(dsm_state=state_of_dsm_val, timestamp=time_val))
    adas_msg.append(adas_msg_parser.create_fcw_can_msg(fcw_state=state_of_fcw_val, object_type=object_type_val, ttc=ttc_val, timestamp=time_val))
    adas_msg.append(adas_msg_parser.create_ldws_can_msg(ldws_state=state_of_ldws_val, timestamp=time_val))

    for msg in adas_msg:
        print(msg)
        can_channel.send(msg)
    
    adas_msg.clear()
    time.sleep(0.1)
   
    if time_val == 26:
        break
    
```
